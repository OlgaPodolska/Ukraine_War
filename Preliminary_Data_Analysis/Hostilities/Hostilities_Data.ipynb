```python
import pandas as pd
import json
from pandas import json_normalize
data = '''
{
"properties": 
    [
    {"OBJECTID": 1},
    {"Date": "2022/08/04"},
    {"Oblast": "Mykolaivska"},
    {"Rayon": "Bashtanskyi"},
    {"Hromada": "Bashtanska"},
    {"City_Settl": "Yavkyne"},
    {"Location__": null},
    {"City_Set_1": "Явкіно"},
    {"City_Set_2": "Явкино"},
    {"GCA_NGCA__": "GCA"},
    {"Status": "Affected"},
    {"Encircled": null},
    {"Humanitari": "https://t.me/mykolaivskaODA/2012"},
    {"admin1Pcod": "UA48"},
    {"admin2Pcod": "UA4802"},
    {"admin3Pcod": "UA4802001"},
    {"admin4Pcod": "UA4802001025"},
    {"Lat__autom": 47.2774122},
    {"Long__auto": 32.5926428},
    {"Long2": 32.5926428},
    {"Lat2": 47.2774122},
    {"Name": "Yavkyne"}
    ],
"status": ["ok"]

}
'''
print(data)
```

```python
# Use json_normalize() to convert JSON to DataFrame
dict = json.loads(data)
event_df = json_normalize(dict['properties']) 
print(event_df)
```

```python
# Drop the rows where all elements are missing.
event_df.dropna(how='all')
```

```python
# Drop the columns where at least one element is missing.
event_df.dropna(axis='columns')
```

```python
# Keep only the rows with at least 2 non-NA values.
event_df.dropna(thresh=2)
```

```python
# Define in which columns to look for missing values.
event_df.dropna(subset=['OBJECTID', 'Date', 'Oblast', 'Rayon', 'Hromada', 'City_Settl', 'Location__', 'City_Set_1', 'City_Set_2', 'GCA_NGCA__', 'Humanitari', 'admin1Pcod', 'admin2Pcod','admin3Pcod','admin4Pcod','Lat__autom','Long__auto','Long2','Lat2','Name'])
```
